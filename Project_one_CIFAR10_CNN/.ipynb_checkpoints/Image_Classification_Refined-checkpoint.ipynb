### Preprocess the data (prepare it for the model)

We now have a training data (X, y) and a test data (X_test, y_test).

We want to preprocess the data:

-  shuffle the data.

-  split (X,y) into a training-set and a validation-set.

-  normalize the data.

#### shuffle the data
To make sure we avoid order-based bias, we want to shuffle the data samples

```python
# Let's examine the labels:
y[:20]
```

It seems like the images are shuffled, but we will shuffle the data before splitting anyway:

```python
import numpy.random as random
# Generate a random permutation of indices
indices = random.permutation(X.shape[0])
# Shuffle X,y by using the indices we generated
X_shuffled = X[indices]
y_shuffled = y[indices]
```

```python
print(f"The labels of the first 10 image samples in y (before shuffling): {y[:10]}")
print(f"The labels of the first 10 image samples in y_shuffled (y after shuffling): {y_shuffled[:10]}")
```

It seems fine, but let's visualize some image samples and make sure the labels match the images:

```python
# Visualise the data
view_random_images(X_shuffled, y_shuffled, classes_names)
```

Great! It seems just fine.

#### Split (X_shuffled, y_shuffled) into a training-set and a validation-set
We want to split the 50,000 training data images into a trainnig-set (X_train, y_train) and a validation-set (X_val, y_val).

We can do so manually, or we can use sklearn's train_test_split.

This time, we will do it manually: 80% for the training set and 20% for the validation set.

```python
# split the data into a training-set (80%) and a validation-set (20%)
val_size = int(0.2 * len(X_shuffled))  # 20% for the validation-set
X_train, y_train = X_shuffled[:-val_size], y_shuffled[:-val_size]  # training set
X_val, y_val = X_shuffled[-val_size:], y_shuffled[-val_size:]  # validation set
# Print the shapes of the sets:
print(f"Training set:\n Number of images in the set: {len(X_train)}\n X_train shape: {X_train.shape}\n y_train shape: {y_train.shape}\n")
print(f"Validation set:\n Number of images in the set: {len(X_val)}\n X_val shape: {X_val.shape}\n y_val shape: {y_val.shape}")
```

#### Normalize the data

Normalizing the data helps improve the convergence of the CNN by scaling pixel values to a uniform range (e.g., [0, 1]), which reduces the risk of numerical instability and allows the model to learn more effectively.

First, let's explore the samples' value ranges:

```python
# Max and Min values:
X_train.max(), X_train.min()
```

We can see that the pixels of our images range between 0 and 255.

### **Model_1: Initial Straightforward Model**

Given the relatively small image size (32x32x3) and moderate complexity of CIFAR-10, a straightforward CNN architecture is sufficient to achieve decent performance without adding unnecessary complexity. The first chosen architecture ('model_1') balances efficiency and accuracy, making it well-suited for this task.

#### Model_1 Architecture

- **Input layer:**
  The Input Layer defines the shape of the input data (32, 32, 3), corresponding to the dimensions of the CIFAR-10 image samples: 32x32 pixels with 3 color channels (RGB). This ensures the model receives data in the expected format.


- **2D-Convolution Layers:*
*  
  Feature extraction begins with Conv2D layers, using 16 to 64 filters, a kernel size of 3, `'same'` padding, and `'relu'` activation function. The kernel size captures spatial patterns, while the `'same'` padding ensures the output dimensions remain consistent after convolution.
  

- **MaxPooling Layers:*
*  
   Each convolutional layer is followed by a MaxPooling layer with `pool_size=2` to reduce spatial dimensions and control overfitting.


- **Flatten Layer:*
*  
  A Flatten layer converts the 2D feature maps into a 1D vector, making them suitable for fully connected layers.


- **Dense Layer:*
*  
  A Dense layer with 128 neurons and `'relu'` activation combines the extracted features and introduces non-linearity.


- **Dropout Layer:*
*  
  A Dropout layer with a rate of 0.5 is applied to reduce overfitting and improve generalization. During training, it randomly deactivates neurons to prevent reliance on specific pathways. During inference (testing), dropout is disabled to ensure the full model capacity is used.


- **Output Dense Layer:*
*  
  The final Dense layer contains 10 neurons (one for each class) with a `'softmax'` activation function, converting the output into class probabilities for multi-class classification.

  > ****Note**:*
* For this project, we will be using a Sequential API. In future projects, where tasks may require more flexibility, we might explore the Functional API.



###### Why This Architecture Works for CIFAR-10

- The dataset's small images and moderate complexity allow this relatively shallow CNN to extract features effectively without excessive computational overhead.

- The increasing number of filters (from 32 to 64) helps capture progressively complex spatial hierarchies.

- The Dropout helps the model generalize well, and reduces the chance of overfitting.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
```

```python
model_1 = Sequential([
    # Input layer
    Input(shape=(32, 32, 3), name="Input layer"),
    # 3 Convolution layers, each followed by MaxPooling layer
    Conv2D(filters=16, kernel_size=(3,3), activation='relu', padding='same', name='Conv_1'),
    MaxPooling2D(pool_size=(2,2), name='MaxPooling_1'),
    Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding='same', name='Conv_2'),
    MaxPooling2D(pool_size=(2,2), name='MaxPooling_2'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='same', name='Conv_3'),
    MaxPooling2D(pool_size=(2,2), name='MaxPooling_3'),
    # Flatten and Dense (Fully-connected) layers
    Flatten(name='Flatten'),
    Dense(units=128, activation='relu', name='Dense_1'),
    Dropout(rate=0.5, name='Dropout'),
    Dense(units=10, activation='softmax', name='Dense_2')
], name="model_1")
```

Now, lets take a look at our model summary to understand its architecture and parameters.

```python
model_1.summary()
```

###### Compile Model_1

Let's compile our model using `sparse_categorical_crossentropy` as the loss function, which is suitable for multi-class classification with integer labels, and the `Adam` optimizer for efficient and adaptive gradient-based optimization. Additionally, we include `metrics=['accuracy']` to track the model's accuracy during training and validation.

```python
model_1.compile(loss=tf.keras.losses.sparse_categorical_crossentropy,
              optimizer=tf.keras.optimizers.Adam(),
             metrics=['accuracy'])
```

###### Hardware Setup

Before training the model, it's crucial to verify that we are utilizing the GPU for computation. GPUs significantly accelerate the training process, especially for deep learning tasks that involve large datasets and complex models.

#### Training

We will train model_1 for 20 `epochs` with a `batch_size` of 64. The validation set will be used to monitor and evaluate the training progress, helping us track the model's performance and detect overfitting.

```python
model_1_history = model_1.fit(x=X_train_norm,
                          y=y_train,
                          batch_size=64,
                          epochs=20,
                          validation_data=(X_val_norm, y_val),
                         )
```

#### Model Evaluation

In this section, we evaluate the performance of model_1 using the following metrics:


- **Accuracy**: Measures the overall correctness of the model's predictions.  

- **Precision**: Evaluates how many of the predicted positive results are true positives.  

- **Recall**: Assesses how well the model identifies all true positive cases.  

- **F1-Score**: A harmonic mean of precision and recall, balancing the two metrics.  

```python
# model predictions
y_pred_probs = model_1.predict(X_val_norm)  
# The model outputs a probability distribution for each sample (10 probabilities, one for each class).
y_pred = tf.argmax(y_pred_probs, axis=1)  
# Converts the probabilities into class predictions by selecting the index with the highest probability.
# y_pred[:5]
```

### **Model_2: Enhanced Architecture**

Based on the analysis of Model_1, we aim to address overfitting, improve its performance, and achieve better results. To do so, we have enhanced the architecture by incorporating the following changes:


- **Additional Filters**:  
    
- Adding more filters to each convolutional layer allows the network to capture more complex and detailed features as the spatial dimensions reduce throughout the layers.  
    
- Using higher filter sizes (**32-64-64**, compared to **16-32-64*
* in Model_1) enhances the network's ability to extract complex patterns and variations, which is particularly beneficial for datasets like CIFAR-10 with diverse and detailed classes.
 

- **Additional Dense and Dropout Layers**:  
    
- To increase the models capacity to learn hierarchical and complex feature representations, the model now includes **3 dense layers*
* (instead of 2 in Model_1):
        
- **Dense Layer 1**: 
            
- 256 units with `'relu'` activation.  
            
- Followed by a dropout layer with `rate=0.5`.  
        
- **Dense Layer 2**: 
            
- 128 units with `'relu'` activation.  
            
- Followed by a dropout layer with `rate=0.3`.  
        
- **Dense Layer 3 (Output)**: 
            
- 10 units with `'softmax'` activation (unchanged from Model_1).  
    
- This combination improves the model's learning capacity while mitigating overfitting, resulting in better generalization to unseen data.  

> ****Note****: Expanding the network's capacity (i.e., adding more parameters) must be approached with caution to avoid overfitting.


- **Batch Normalization (BN)**:  
    
- Batch Normalization layers are added after each convolutional layer to stabilize and accelerate training by normalizing activations within each batch, making the network less sensitive to initialization and learning rate selection.
 

- **Increased Batch Size**:  
    
- The batch size was increased from **64 to 256**, reducing fluctuations in validation loss and accuracy curves by stabilizing gradient updates. This adjustment might leed to smoother and more consistent performance during training.


- **Early Stopping Mechanism**:  
    
- Early stopping is implemented to monitor validation loss during training and halt training when no improvement is observed for a predefined number of epochs, preventing the model from overfitting.  


- **Learning Rate Scheduler**:  
    
- A learning rate scheduler dynamically adjusts the learning rate during training, enabling optimal convergence by reducing the learning rate as the model approaches a minimum loss.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
```

```python
model_2 = Sequential([
    # Input layer
    Input(shape=(32, 32, 3), name="Input layer"),
    # 3 Convolution layers, each followed by MaxPooling layer
    Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding='same', name='Conv_1'),
    BatchNormalization(name='BN_1'),  # Added Batch Normalization
    MaxPooling2D(pool_size=(2,2), name='MaxPooling_1'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='same', name='Conv_2'),
    BatchNormalization(name='BN_2'),  # Added Batch Normalization
    MaxPooling2D(pool_size=(2,2), name='MaxPooling_2'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='same', name='Conv_3'),
    BatchNormalization(name='BN_3'),  # Added Batch Normalization
    MaxPooling2D(pool_size=(2,2), name='MaxPooling_3'),
    # Flatten and Dense (Fully-connected) layers
    Flatten(name='Flatten'),
    Dense(units=256, activation='relu', name='Dense_1'),
    Dropout(rate=0.5, name='Dropout_1'),
    Dense(units=128, activation='relu', name='Dense_2'),
    Dropout(rate=0.3, name='Dropout_2'),
    Dense(units=10, activation='softmax', name='Dense_3')
], name="model_2")
```

Now, lets take a look at our model summary to understand its architecture and parameters.

```python
model_2.summary()
```

****Note**:*
*

###### Callbacks
Lets define the Early Stopping and Learning Rate Scheduler callbacks:

```python
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
from tensorflow.keras.optimizers import Adam
from helper_functions import lr_schedule
# LearningRateScheduler callback
# define Learning Rate Scheduler Callback
lr_scheduler = LearningRateScheduler(
    schedule=lr_schedule,     # use the learning rate schedule function from helper_functions
    verbose=0                 # print the learning rate at each epoch
)
# EarlyStopping callback
# define Early Stopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',       # monitor validation loss
    patience=6,               # stop training if no improvement after 6 epochs
    restore_best_weights=True # restore the best weights after stopping
)
```

###### Compile Model_2

```python
# Compile the model
model_2.compile(
    optimizer=Adam(learning_rate=1e-3),  # Initial learning rate
    loss='sparse_categorical_crossentropy',  # Loss function for multi-class classification
    metrics=['accuracy']  # Evaluation metrics
)
```

###### Hardware Setup

Before training the model, it's crucial to verify that we are utilizing the GPU for computation. GPUs significantly accelerate the training process, especially for deep learning tasks that involve large datasets and complex models. 

The following code ensures that a GPU is detected and will be used:

```python
gpu_devices = tf.config.list_physical_devices('GPU')
print("Training will be performed on:", "GPU" if gpu_devices else "CPU")
```

#### Training

We will train model_2 for 20 `epochs` with a `batch_size` of 64. The validation set will be used to monitor and evaluate the training progress, helping us track the model's performance and detect overfitting.

We use early_stopping and lr_scheduler callbacks.

```python
# Train the model
model_2_history = model_2.fit(
    X_train_norm, 
    y_train,
    validation_data=(X_val_norm, y_val),  # Validation set for monitoring performance
    epochs=20,                      # Maximum number of epochs
    batch_size=256,                  # Batch size
    callbacks=[early_stopping, lr_scheduler],  # Add callbacks here
)
```

In this section, we evaluate the performance of model_2 using the following metrics:


- **Accuracy**: Measures the overall correctness of the model's predictions.  

- **Precision**: Evaluates how many of the predicted positive results are true positives.  

- **Recall**: Assesses how well the model identifies all true positive cases.  

- **F1-Score**: A harmonic mean of precision and recall, balancing the two metrics.  

```python
# model predictions
y_pred_probs = model_2.predict(X_val_norm)  
# The model outputs a probability distribution for each sample (10 probabilities, one for each class).
y_pred = tf.argmax(y_pred_probs, axis=1)  
# Converts the probabilities into class predictions by selecting the index with the highest probability.
```

```python
# checking y_val and y_pred shapes
y_val.shape, y_pred.shape
```

```python
# Expand the dimensions of `y_pred` to match the required shape for evaluation
y_pred_fixed = np.expand_dims(y_pred, axis=1)  # Adds an extra dimension at axis=1, converting shape from (n_samples,) to (n_samples, 1)
y_pred_fixed.shape  # Check the updated shape
```
